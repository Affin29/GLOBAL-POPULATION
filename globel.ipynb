```python
plt.figure(figsize=(8,5))
sns.histplot(pf['Population(in millions)'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Total Population (in millions)")
plt.xlabel("Population (millions)")
plt.ylabel("Frequency")
plt.show()
```

Distribution of Total Population

```python
top10_pop = pf.nlargest(10, 'Population(in millions)')
plt.figure(figsize=(10,6))
sns.barplot(data=top10_pop, x='Population(in millions)', y='Country', palette='viridis')
plt.title("Top 10 Most Populous Countries (2024)")
plt.xlabel("Population (millions)")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=pf, x='Population(in millions)', y='Population density',
                hue='Population Aged 0 to 14 (%)', size='Population Aged 60 and Over (%)', sizes=(20,200))
plt.title("Population vs Population Density (colored by age structure)")
plt.xlabel("Population (millions)")
plt.ylabel("Population Density")
plt.legend(title="Age 0â€“14 (%)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=pf, x='Population Aged 0 to 14 (%)', y='Population Aged 60 and Over (%)',
                hue='Population(in millions)', size='Population density', sizes=(20,200))
plt.title("Youth vs Elderly Population (%)")
plt.xlabel("Population Aged 0â€“14 (%)")
plt.ylabel("Population Aged 60+ (%)")
plt.legend(title="Total Population (millions)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pf.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Global Population Statistics")
plt.show()
```

```python
top10_density = pf.nlargest(10, 'Population density')

plt.figure(figsize=(10,6))
sns.barplot(data=top10_density,
            x='Population density',
            y='Country',
            palette='magma')
plt.title("Top 10 Densely Populated Countries")
plt.xlabel("Population Density (people per sq. km)")
plt.ylabel("Country")
plt.show()
```
